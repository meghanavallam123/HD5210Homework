## Step 1: Calculate PRIEST Clinical Severity Score

This scoring algorithm can be found [here on the MDCalc website](https://www.mdcalc.com/priest-covid-19-clinical-severity-score#evidence).  

1. You will need to write a function called **priest()** with the following input parameters.  
 * Sex (Gender assigned at birth)
 * Age in years
 * Respiratory rate in breaths per minute
 * Oxygen saturation as a percent between 0 and 1
 * Heart rate in beats per minute
 * Systolic BP in mmHg
 * Temperature in degrees C
 * Alertness as a string description
 * Inspired Oxygen as as string description
 * Performance Status as a string description
2. The function will need to follow the algorithm provided on the MDCalc website to compute a risk percentage that should be returned as a numeric value between 0 and 1.
3. Be sure to use docstring documentation and at least three built-in docstring test cases.
4. Assume that the input values that are strings could be any combination of upper or lower case. For example: 'male', 'Male', 'MALE', 'MalE' should all be interpretted by your code as male.
5. If any of the inputs are invalid (for example a sex value that is not recognizable as male or female) your code should return None.

NOTES:
1. In the final step there is a table that translates from **PRIEST Score** to **30-day probability of an outcome** but the last two probabilities are shown as ranges (59-88% and >99%).  Our code needs to output a single number, however. For our code, use the following rule:
 * If PRIEST score is between 17 and 25, the probability you return should be 0.59
 * If PRIEST score is greater than or equal to 26, the probability you return should be 0.99


````python
def priest(pat_gender, pat_age, pat_resp_rate, pat_sat, pat_heart, pat_bp, pat_temperature, pat_alert, pat_ins_ox, pat_status):

    '''
        (str, int, int, float, int, int, float, str, str, str) -> (float)
        This function returns the risk percentage, represented as the "priest score", for assessing the severity of COVID-19 clinical symptoms.

        It takes multiple input parameters such as gender, age, respiratory rate, oxygen saturation, heart rate, blood pressure, body temperature, alertness, inspired oxygen support, and performance status.

        The output of the function is a risk percentage as a decimal value within the range of 0 to 1. If any of the input parameters are invalid, the function will return a None value.


        Test Cases -->

        >>> priest("Male",18,8.0,96,105.0,115.0,38.1,"ALERT","AIR","limited strenuous activity, can do light activity")
        0.18

        >>> priest("MALE",21,22.0,96,50.0,115.0,38.1,"ALERT","AIR","limited strenuous activity, can do light activity")
        0.15

        >>> priest("female",19,22.0,96,105.0,115.0,38.1,"confused or not alert","supplemental oxygen","unrestricted normal activity")
        0.26

    '''



    # Initialize the pat_count to 0
    pat_count = 0



    # Convert string variable to lowercase

    pat_gender = pat_gender.lower()
    pat_alert = pat_alert.lower()
    pat_ins_ox = pat_ins_ox.lower()
    pat_status = pat_status.lower()




    # Check for correct input


    if pat_gender not in ('male', 'female'):
        return None

    if pat_alert not in ('alert', 'confused or not alert'):
        return None

    if pat_ins_ox not in ('air', 'supplemental oxygen'):
        return None

    if pat_status not in ('unrestricted normal activity', 'limited strenuous activity, can do light activity', 'limited activity, can self-care', 'limited self-care', 'bed/chair bound, no self-care'):
        return None




    # compute pat_count

    # update the value of pat_count for pat_gender
    if(pat_gender == 'male'):
        pat_count +=1


    # update the value of pat_count for pat_age
    if(pat_age >= 50 and pat_age <= 65):
        pat_count += 2
    elif(pat_age >= 66 and pat_age <= 80):
        pat_count += 3
    elif(pat_age > 80):
        pat_count += 4


    # update the value of pat_count for respiratory rate
    if(pat_resp_rate < 9):
        pat_count += 3
    elif(pat_resp_rate >= 9 and pat_resp_rate <= 11):
        pat_count += 1
    elif(pat_resp_rate >= 12 and pat_resp_rate <= 20):
        pat_count += 0
    elif(pat_resp_rate >= 21 and pat_resp_rate <= 24):
        pat_count += 2
    elif(pat_resp_rate > 24):
        pat_count += 3


    # update the value of pat_count for oxygen saturation
    if(pat_sat > 95):
        pat_count += 0
    elif(pat_sat >= 94 and pat_sat <= 95):
        pat_count += 1
    elif(pat_sat >= 92 and pat_sat <= 93):
        pat_count += 2
    elif(pat_sat < 92):
        pat_count += 3


    # update the value of pat_count for pat_heart rate
    if(pat_heart < 41):
        pat_count += 3
    elif(pat_heart >= 41 and pat_heart <= 50):
        pat_count += 1
    elif(pat_heart >= 51 and pat_heart <= 90):
        pat_count += 0
    elif(pat_heart >= 91 and pat_heart <= 110):
        pat_count += 1
    elif(pat_heart >= 111 and pat_heart <= 130):
        pat_count += 2
    elif(pat_heart > 130):
        pat_count += 3


    # pat_count for systolic blood pressure
    if(pat_bp < 91):
        pat_count += 3
    elif(pat_bp >= 91 and pat_bp <= 100):
        pat_count += 2
    elif(pat_bp >= 101 and pat_bp <= 110):
        pat_count += 1
    elif(pat_bp >= 111 and pat_bp <= 219):
        pat_count += 0
    elif( pat_bp > 219):
        pat_count += 3


    # update the value of pat_count for body pat_temperature
    if(pat_temperature < 35.1):
        pat_count += 3
    elif(pat_temperature >= 35.1 and pat_temperature <= 36):
        pat_count += 1
    elif(pat_temperature >= 36.1 and pat_temperature <= 38):
        pat_count += 0
    elif(pat_temperature >= 38.1 and pat_temperature <= 39):
        pat_count += 1
    elif( pat_temperature > 39):
        pat_count += 2


    # update the value of pat_count for pat_alert
    if(pat_alert == "alert"):
        pat_count += 0
    elif(pat_alert == "confused or not alert"):
        pat_count += 3


    # update the value of pat_count for pat_ins_ox
    if(pat_ins_ox == "air"):
        pat_count += 0
    elif(pat_ins_ox == "supplemental oxygen"):
        pat_count += 2


    # update the value of pat_count for pat_status
    if(pat_status =="unrestricted normal activity"):
        pat_count += 0
    elif(pat_status =="limited strenuous activity, can do light activity"):
        pat_count += 1
    elif(pat_status=="limited activity, can self-care"):
        pat_count += 2
    elif(pat_status=="limited self-care"):
        pat_count += 3
    elif(pat_status=="bed/chair bound, no self-care"):
        pat_count += 4


    # compute the PRIEST SCORE
    priest_score = 0

    if(pat_count == 0 or pat_count == 1):
        priest_score=0.01
    elif(pat_count == 2 or pat_count == 3):
        priest_score=0.02
    elif(pat_count==4):
        priest_score=0.03
    elif(pat_count==5):
        priest_score=0.09
    elif(pat_count==6):
        priest_score=0.15
    elif(pat_count==7):
        priest_score=0.18
    elif(pat_count==8):
        priest_score=0.22
    elif(pat_count==9):
        priest_score=0.26
    elif(pat_count==10):
        priest_score=0.29
    elif(pat_count==11):
        priest_score=0.34
    elif(pat_count==12):
        priest_score=0.38
    elif(pat_count==13):
        priest_score=0.46
    elif(pat_count==14):
        priest_score=0.47
    elif(pat_count==15):
        priest_score=0.49
    elif(pat_count==16):
        priest_score=0.55
    elif(pat_count>=17 and pat_count<=25):
        priest_score=0.59
    elif(pat_count>=26):
        priest_score=0.99

    return priest_score
````

## Part 2: Find a hospital

The next thing we have to do is figure out where to send this particular patient.  The guidelines on where to send a patient are based on their age (pediatric, adult, geriatric), sex, and risk percentage.  Luckily, you don't have to implement these rules. I already have. All you have to do is use a REST web service that I've created for you.

You'll want to use Python to make a call to my REST web service similar to the example URL below. The first part of the URL will be the same for everyone and every request that you make. What you will need to modify for each of your requests is the information after the question mark.

```
https://oumdj6oci2.execute-api.us-east-1.amazonaws.com/prd/?age=40&sex=male&risk_pct=0.1
```

The example above asks my web service where a 40-year old male with a risk of 10% should go.  What the web service will return back is a JSON string containing the information you need.  That JSON will look like this:

```json
{
  "age": "40",
  "sex": "male",
  "risk": "0.1",
  "hospital": "Southwest Hospital and Medical Center"
}
```

My function is not smart enough to understand `'MALE'` is the same as `'male'`.  You have to send it exactly `'male'` or `'female'`

1. Your job is to write a function called **find_hospital()** that takes age, sex, and risk as parameters.
2. Your function should call this REST web service using the `requests` module
3. Then your function will need to interpret the JSON it gets and return just the name of the hospital
4. If anything fails, return None
5. Include a good docstring with at least three test cases.


````python
import requests
import json

def find_hospital(pat_age, pat_sex, priest_score):

    '''
         (string, string, float) -> String
        This function takes "patient's age, gender, and priest score" as input parameters and returns the name of the hospital as the output.

        >>> find_hospital(40, 'male', 0.02)
        'Southwest Hospital and Medical Center'

        >>> find_hospital(48,'female',0.46)
        'Emory Dunwoody Medical Center'

        >>> find_hospital(19, 'MALE', 0.46)
        'Emory Dunwoody Medical Center'

    '''

    try:

        #gender to lowercase
        pat_sex = pat_sex.lower()

        pat_age = int(pat_age)

        given_url = f'https://oumdj6oci2.execute-api.us-east-1.amazonaws.com/prd/?age={pat_age}&sex={pat_sex}&risk_pct={priest_score}'

        my_response = requests.get(given_url)

        if my_response.status_code == 200:

            content = my_response.content

            dc_content = content.decode('utf-8')

            d_json_data = json.loads(dc_content)

            main_op = d_json_data['hospital']

            return main_op

    except:

        return None
````

## Part 4: Run the risk calculator on a population

Your Python program should use your code above to process all of these rows, determine the hospital and address, and return a list whose items are a dictionary like this: `{ patient_number: [sex, age, breath, o2sat, heart, systolic, temp, alertness, inspired, status, hospital, address]}`.
````python
import csv

def process_people(given_url):

    '''
         (String) -> Dictionary
        This function fetches patient information from a given URL and apply obtained info to the previously mentioned functions to this data, and generates a dictionary containing patient specifics.

        Each patient is distinctly recognized by their unique number, which serves as the dictionary key.

    '''

    # Send a GET request
    my_response = requests.get(given_url)

    content = my_response.content

    dc_content = content.decode('utf-8')

    # Split the string content into lines and remove the header row
    l = dc_content.split('\n')

    l.pop(0)

    # Create a CSV reader to process the data
    main_list = csv.reader(l, delimiter='|')

    ret_fn_op = {}

    for item in main_list:

        if len(item) == 11:

            id = item[0]
            pat_gender = item[1]
            pat_age = int(item[2])
            pat_resp_rate = int(item[3])
            pat_sat = float(item[4])* 100
            pat_heart = int(item[5])
            pat_bp = int(item[6])
            pat_temperature = float(item[7])
            pat_alert = item[8]
            pat_ins_ox = item[9]
            pat_status = item[10]

            priest_score = priest(pat_gender, pat_age, pat_resp_rate, pat_sat, pat_heart, pat_bp, pat_temperature, pat_alert, pat_ins_ox, pat_status)
            hospital_name = find_hospital(pat_age, pat_gender, priest_score)
            hospital_address = get_address(hospital_name)

            ret_fn_op[id] = [pat_gender, pat_age, pat_resp_rate, pat_sat/100, pat_heart, pat_bp, pat_temperature, pat_alert, pat_ins_ox, pat_status, priest_score,hospital_name,hospital_address]

    return ret_fn_op

patient_dict = process_people('https://drive.google.com/uc?export=download&id=1fLxJN9YGUqmqExrilxSS8furwUER5HHh')
````
